<H1>Introduction</H1>
The problem that we plan to solve in this analysis is finding the best location for one or two coffee shops for an entrepreneur who is contemplating opening a coffee shop in the Milwaukee Wisconsin area. She already has several coffee shops in New Orleans and Philadelphia. The client wanted a fairly quick and low cost analysis to determine if there is enough potential for a successful new coffee shop. If this analysis shows that potential, she plans to fly out to Milwaukee and perform due diligence on the coffee shops this analysis presents and on the selected areas.
She has learned from the years of experience in her other coffee shops that her shops are most successful when a new location:<P>
<UL><LI>Is not more than two miles from the center of town; </LI>
<LI>Is within walking distance of at least two museums;</LI>
<LI>Has no more than two coffee shops in a two block radius;</LI>
<LI>Has one or more bars on the same block.</LI></UL>
<H1>Data</H1>
Because the client already has an  idea of the ideal business environment for her potential coffee shops, this analysis combines three different FourSquare data sets:<P>
1.	Venues search for museums in the Milwaukee area<BR>
2.	Venues search for coffee shops in the Milwaukee area<BR>
3.	Venues search for bars  in the Milwaukee area<BR><P>
Each data set is pulled from FourSquares’ venue data and has a similar structure, with the key fields being the venue category, venue name, and location. 
<H1>Methodology</H1>
At our client’s request, initially the data locations for museums was pulled so we could see the breadth of this type of venue in Milwaukee. We first had a limit of twenty museums within 2000 meters of Milwaukee’s center, because we didn’t want to overwhelm the client with data, and she wanted  to stay close to downtown. We noticed that some listings were near duplicates, for example, a special exhibit at an existing museum might be listed a separate museum. Because of the potential traffic of a special exhibit, we decided to keep these types of data separate. After reviewing the data with our client to determine the best next steps, we decided to expand the limit to 50 museums within 5000 meters to see the full scope of this venue. The data showed us there are 30 museums in Milwaukee that fit those criteria. 
In order to best match the coffee shop data to the museum we used the same radius and limit criteria. We also examined the data for each coffee shop and saw two that were marginally relevant, such as a coffee services in a grocery store and in a church. But without an actual visit to those locations we decided to retain all coffee shop data for this initial analysis.
After the initial approach to reviewing the data, the main method of analysis used a plot that contains several features:<P>
<OL><LI>Plotted circle at the center of Milwaukee</LI>
<LI>Plotted circles for museums</LI>
<LI>Plotted circles for coffee shops</LI>
<LI>Plotted circles for bars</LI>
</OL>
<P>Because the plot can be zoomed in, the client can find the exact street and building of suitable locations that meet her business parameters. The plot is very easy for her to read and gave her more insight into the area than she was expecting. Because the results also included  easy-to-read tables of both the museum and coffee shops, she could see synergies or potential specializations. For example the low number of coffee shops near the Harley Davidson Museum  might allow her to collaborate with the company and  sell Harley Davidson coffee cups. The existence of two Jewish themed museums might allow her to specialize to Jewish clientele. 
The other analytical approach  that was useful, and facilitated by the datasets being fairly small, were the tables of results from the Foursquare venue data. As mentioned above this data served as information on the types of museums and coffee shops so the client was able to get a good understanding of the different venues.
<H1>Results</H1>
The main goal of this analysis using FourSquare data was to show the client the level; of potential for her to open one or more new coffee shops. The analysis was based on her prior professional experience that locations near museums were the most lucrative, and that having a bar nearby added to the potential value of a location. On the other hand, too much competition made a location less lucrative. As show in the maps below, especially Figure 2, there are two locations near the center of Milwaukee that have more than one museum, have at least one bar and have only one coffee shop.

 <H1>Conclusion</H1>
Although this was a relatively simple analysis that didn’t require regression or machine learning, it was an important and low cost first step for the client. The analysis showed one location for a coffee shop with very high potential. It met the client’s expectations of a good location. The second most lucrative neighborhood was not ideal, but it may have high quality niche clientele, that was of interest to the client. So the data collection and methods used to represent the data were very successful.<P><P>
There are some weaknesses with this analysis, because it did not analyze restaurants that could also act as a bar. As well some of the table data produced from FourSquare did not have the venue’s address, making it somewhat inconvenient for the client to perform due diligence on that venue, although she only needs to google the venue’s address. Still it would be ideal for this have complete data.
Another takeaway from this analysis is that with very little effort, she or her analyst can use the code provided and simply swap out a new city’s longitude and latitude. Now that the analysis is build, she can peruse other cities’ potential in minutes. This again, makes this simple data mapping project a very affordable first step for someone looking to open a coffee shop in any city.
